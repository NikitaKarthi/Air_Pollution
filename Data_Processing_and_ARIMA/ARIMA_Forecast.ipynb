# PM25

## Predict

```r
Jan21PM25_fit <- auto.arima(ts_Jan21PM25_Stationary_value, seasonal = FALSE)

X_testPM25 <- read.csv("ARIMA_TS_TrainTestValid/X_validPM25.csv", header = TRUE, sep = ",")
y_testPM25 <- read.csv('ARIMA_TS_TrainTestValid/y_validPM25.csv', header = TRUE, sep = ",")  
forecast_list <- list()

for (x in 1:(nrow(X_testPM25))){
    train_ts = ts(unlist(X_testPM25[1:x,2:6]))
    forecast_ts <- forecast(train_ts, model = Jan21PM25_fit, h=1)
    forecast_mean= forecast_ts['mean']    
    forecast_list <- append(forecast_list, unlist(forecast_mean))
    #print(unlist(forecast_mean))
}

Jan21PM25_predict_valid <- as.data.frame(t(as.data.frame(forecast_list)))

write.csv(Jan21PM25_predict_valid, 'ARIMA_TS_TrainTestValid/Jan21PM25_predict_Valid.csv')
```

```r
forecastList <- list()

for (i in 1:length(forecast_list)){
    temp <- as.numeric(forecast_list[i])
    forecastList <- append(forecastList, temp)   
    #print(temp)
} 

plot(c(1:784), forecast_list, type = 'l',xlab="Observation Number", ylab = 'Pollution Level', col = "blue")
lines(c(1:784), as.list(y_testPM25[,2]), xlab="Observation Number", type = 'l', ylab = 'Pollution Level', col = "red")

legend("topright", c("forecastPM25", "real_valuePM25"), fill= c("blue", "red"))
```
